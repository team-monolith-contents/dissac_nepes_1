### 1차원 리스트 익히기

**리스트란?**

* 즉, 여러 데이터들을 묶어서 사용해야할 때 하나의 이름으로 묶어주는 방법입니다. 
* 리스트 안의 데이터들을 요소(element)라고 합니다. 
* 리스트는 다음과 같이 대괄호 안에 자료들을 입력함으로써 선언됩니다. 각 자료는 쉼표로 구분해주어야 합니다.   
> 리스트 이름 = [데이터 1, 데이터 2, ... ]  
예시 : a = [1,2,3]

* 빈 리스트는 대괄호 안을 비운 채로 작성해주면 됩니다. 
> 리스트 이름 = []

혹시 **문제 2.** 에서 입력값에 0이나 1이 아닌 다른 값을 입력해보셨나요? 아마 에러가 떴을 겁니다.

이번에는 조건문을 사용하여 입력을 검증하는 방법을 알아보겠습니다.

### 조건문

**조건문이란?**  
* 조건에 따라 다르게 동작을 실행할 수 있도록 작성한 코드입니다.
* 조건문을 사용해, 조건이 참인 경우와 거짓인 경우를 나누어 각각 다른 코드가 실행되도록 명령할 수 있습니다.  

코드는 다음과 같이 여러 부분으로 쪼개볼 수 있어요.
1. 비교문
    
    `a > b`, `b < a`와 같은 구문들은 수학시간에 배웠죠? 바로 a가 b보다 크냐 작냐를 판정하는 비교문이에요. 비교문에 쓰이는 `>`, `<`와 같은 것을 비교연산자라고 해요.
    비교 연산자에는 여러 종류가 있어요.  
    **비교 연산자의 종류**
    비교 연산자의 종류 | 설명 | 예시 | True of False
    ---|:---:|---|---
    a>b | a가 b 초과일 때 참 | 17>5 | True
    a>=b | a가 b 이상일 때 참 | 2>=4 | False
    a==b | a와 b가 동일할 때 참 | 3==3 |True
    a!=b | a와 b가 다른 경우 참 | 7!=7 | False
    a<b | a가 b 미만일 때 참 | 3<4 |True
    a<=b | a가 b 이하일 때 참 | 7<=9 | True

    **(주의) a와 b가 동일함을 나타내는 연산자 == 를 변수에 값을 대입해주는 대입 연산자 = 과 헷갈리지 않게 주의해야 합니다.**  
    - a == b : a와 b가 동일할 때 참을 반환합니다.   
    a와 b는 모두 값입니다. 따라서 둘 다 이전에 선언되지 않았다면 오류가 발생합니다.    
    - a = b : 변수 a에 값 b를 대입해줍니다.   
    a는 변수로서 기능하지만, b는 a에 저장될 값으로서 기능합니다. 
    따라서 이 경우 b가 이 코드 이전에 선언되지 않은 경우 오류가 발생합니다. 

    **(주의) >=, <=, != 에서 "="는 반드시 오른쪽에 작성되어야 합니다. (=>, =<, =!와 같이 작성하지 않도록 주의하세요.)**

2. 조건문
    
    조건을 만족하는 경우, 그 안에 있는 코드들이 실행됩니다. 예를 들어 아래 예시는 `a가 b보다 큽니다`가 출력될 겁니다.
      
    반대로 a가 b보다 작은 경우라면, 아래와 같이 `a가 b보다 작습니다`가 출력될 겁니다.  
      
    만약 크지도 작지도 않은 같은 경우라면, 위와 같이 `a와 b가 같습니다`가 출력될 겁니다.

### 2차원 리스트

**2차원 리스트란**  
앞서 배웠듯이, 요소로 다시 1차원 리스트를 가지는 경우, 이를 **2차원 리스트**라고 부릅니다.  
  

변수 b는 각 요소로 1차원 리스트인 `[1, 2]`, `[3, 4]`, `[5, 6]`을 원소로 가지기 때문에 2차원 리스트입니다.  
변수 b의 첫번째 요소는 `[1, 2]`이고, 두번째 요소는 `[3, 4]`, 세번째 요소는 `[5, 6]`입니다.  
우리는 이를 아래 코드에서도 확인할 수 있습니다.

````python
b = [[1, 2], [3, 4], [5, 6]]
print(b[0])
print(b[1])
print(b[2])
````

**2차원 리스트 반복문**  
위처럼 인덱스를 활용하면 각 리스트의 요소에 접근할 수 있지만, 이 경우 2차원 리스트의 길이를 알아야 합니다.  
Python의 반복문인 for문을 활용하면 2차원 리스트의 각 요소인 1차원 리스트에 접근할 수 있습니다.

**반복문이란?**  
* 단순히 같은 작업을 여러 번 반복해야 할 때, 반복문을 활용하면 시간 소모를 줄일 수 있습니다. 
* 반복문은 같은 코드를 여러 번 입력하지 않고도 같은 코드를 여러 번 반복하여 실행시킬 수 있는 방법입니다.
* 기본적으로 for를 사용한 반복문과 while을 사용한 반복문이 사용됩니다.



위 그림은 for문의 문법을 나타냅니다. 리스트의 첫번째 요소부터 마지막 요소까지 변수에 차례대로 대입해 코드1을 실행합니다.  
아래는 그 예시입니다. 아래 코드의 출력 결과는 위에 있는 예제와 같습니다.

````python
arr = [[1, 2], [3, 4], [5, 6]]
for item in arr:
    print(item)
````
