### 마을 구조 파악하기
먼저 철수가 우편 배달을 하고 있는 마을의 구조를 알아 봅시다.  
1. 전체 도로 구조
* 철수의 마을의 입구에는 여러개의 갈림길이 있습니다. 갈림길을 지날수록, 도달할 수 있는 집의 갯수는 반씩 줄어듭니다.
  
2. 갈림길
* 갈림길의 왼쪽으로 가면 1부터 8의 주소를 가진 집에 갈 수 있고, 오른쪽으로 가면 9부터 16의 주소를 가진 집에 갈 수 있습니다.
* 집의 주소가 나뉘는 값은 가장 왼쪽에 있는 집의 주소와 오른쪽에 있는 집의 주소의 평균입니다.
    * 예를 들어, 1과 16의 평균인 8.5는 8보다 크고 9보다 작습니다.
    * 따라서 갈림길에서 왼쪽으로 가면 8.5보다 작은 집의 주소에 도달할 수 있고, 오른족으로 가면 8.5보다 큰 집의 주소에 도달할 수 있습니다.

### 일반화시키기
* 위의 2. 갈림길과 집 주소 간의 관계를 1~8의 집이 존재하는 A갈림길에 적용해봅시다.  
  

1과 8의 평균인 4.5는 4보다 크고 5보다 작습니다.

따라서 갈림길에서 왼쪽으로 가면 4.5보다 작은 집의 주소에 도달할 수 있고, 오른족으로 가면 4.5보다 큰 집의 주소에 도달할 수 있습니다.

* 갈림길 A에서 7의 주소를 가지는 집에 도달하기까지의 과정을 갈림길을 기준으로 설명해 봅시다.  
  

1. 오른족으로
    * 주소가 4.5보다 큰 집들에 갈 수 있다. (4.5 < 주소)
2. 오른족으로
    * 주소가 6.5보다 큰 집들에 갈 수 있다. (6.5 < 주소)
3. 왼쪽으로
    * 주소가 7.5보다 작은 집들에 갈 수 있다. (6.5 < 주소 < 7.5)

### 갈 수 있는 주소의 범위 찾기
먼저, 현재 갈림길에서 왼쪽과 오른족으로 갔을 때 갈 수 있는 집주소의 범위를 구해보겠습니다.
1. `low`와 `high`라는 변수를 선언하고 알맞은 값을 저장합니다.
    * `low`는 현재 갈림길에서 갈 수 있는 집의 가장 작은 주소입니다.
    * `high`는 현재 갈림길에서 갈 수 잇는 집의 가장 큰 주소입니다.
2. 갈림길의 왼쪽으로 갔을 때 low와 high가 어떻게 변하는지 출력해봅시다.
3. 갈림길의 오른쪽으로 갔을 때 low와 high가 어떻게 변하는지 출력해봅시다.

힌트
* 갈림길의 값이 정수가 아니라 실수인 경우, 이를 정수로 알맞게 변환해주어야 합니다.
* python에서 실수를 정수로 변환시키는 함수는 다음과 같습니다.
    * `round(a)`: a의 반올림한 값을 반환한다.
    * `math.floor(a)`: a의 내림한 값을 반환한다. `import math`를 먼저 실행해야 한다.
    * `math.ceil(a)`: a의 올림한 값을 반환한다. `import math`를 먼저 실행해야 한다.


```python
low = 1
high = 1024  # 초기값은 전체 주소의 가장 작은 값과 가장 큰 값으로 설정해줍니다.

mid = (low + high) / 2 

# print(최소값, 최대값)의 형태로 작성했습니다. 
# 왼쪽으로 갔을 때
print(low, math.floor(mid)) # 왼쪽으로 간 경우 최대값은 mid를 내려주어야 합니다. 앞선 예시에서 가운데가 8.5일 때 왼쪽의 최대값이 8이었던 것을 생각하면 어렵지 않습니다. 
# 오른쪽으로 갔을 때
print(math.ceil(mid), high) # 오른쪽으로 간 경우의 최소값은 mid를 올려주어야 합니다. 
```

### 어느 갈림길을 선택하지 고르기
자신이 가고 싶은 집의 주소를 알고 있을 때, 어느 갈림길로 가야할지 결정해봅시다.
* `low`는 1, `high`는 8이라 대입합니다.
* 자신이 가고 싶은 집의 주소는 7이라 가정합니다.
* 현재 갈림길에서 어느 방향으로 가는지, `low`와 `high`변수의 값을 출력해 봅시다.

```python
low = 1
high = 8
goal = 7
mid = (low + high) / 2

if (mid < goal): # 갈림길(mid)보다 목표한 집주소(goal)가 크다면 오른쪽으로 갑니다.
    print('오른쪽')
    print(math.ceil(mid), high) # 올림한 값을 최소값으로 설정합니다. 
else: # 아니라면 왼쪽으로 간다.
    print('왼쪽')
    print(low, math.floor(mid)) # 내린 값을 최대값으로 설정합니다. 
```

### 반복문 적용하기
3.2에서 구현한 것을 모든 갈림길에 대해 적용해봅시다.
* `low`는 1, `high`는 1024라 대입합니다.
* 철수가 가고 싶은 집의 주소는 316입니다.
* 각 갈림길마다 어느 방향으로 가야 하는지, `low`와 `high`변수의 값을 출력합니다.

힌트
* 갈림길을 마주치는 횟수를 정하기 힘드므로, `while` 문을 사용하여 동일한 구문을 반복합니다.
* 갈림길을 모두 지났을 때 `low`와 `high`변수가 어떻게 될지 생각해보고, 이를 while문의 탈출 조건문으로 사용해 봅시다.

```python
import math

low = 1
high = 1024
goal = 316

while (low < high):  # low가 high보다 크거나 같을 때 반복문이 중단됩니다. 
    mid = (low + high) / 2
    print(low, high)
    if mid < goal:
        print('오른쪽')
        low = math.ceil(mid)
    elif mid > goal:
        print('왼쪽')
        high = math.floor(mid)

print(low, high) # low가 high가 같아지는 모습을 관찰할 수 있다.
```

### 반복문 횟수 세기
* 반복문이 돌아간 횟수를 센 뒤, 이를 출력해 봅시다.
* `high`의 값에 512, 256, 128, 64를 넣어보며 각 횟수가 어떻게 달라지는지를 관찰해봅시다. 어떤 관계를 찾을 수 있을까요?
* 철수는 제때 물건을 배달할 수 있는지를 판정해보자! (갈림길을 지난후 다음 갈림길 혹은 집을 마주하기까지 1분이 소요된다고 가정합니다.)

```python
import math

low = 1
highs = [512, 256, 128, 64]
goal = 12

cnt = 0

for high in highs:
    print("========", high, "========")
    cnt=0
    while (low < high):
        cnt += 1

        mid = (low + high) / 2
        print(low, high)
        if mid < goal:
            print('오른쪽')
            low = math.ceil(mid)
        elif mid > goal:
            print('왼족')
            high = math.floor(mid)

    print(low, high) # low가 high가 같아지는 모습을 관찰할 수 있다.
    print('갈림길 횟수', cnt) # 2^cnt가 high인 모습을 관찰할 수 있다. (갈림길을 8번 지나기 때문에 목적지까지 걸리는 시간은 8분입니다!)

    if cnt<=10:
        print("\nSafe\n")
    else:
        print("\nFail\n")
```
